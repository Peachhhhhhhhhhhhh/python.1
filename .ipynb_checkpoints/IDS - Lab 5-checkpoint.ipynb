## Data

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
x = np.arange(0,100)
y = x*2
z = x**2
df = pd.read_csv('Superstore.csv',encoding = 'iso-8859-1')
```

## Exercise 1


```python
plt.plot(x,y)
plt.title("Title")
plt.xlabel('x')
plt.ylabel('y')
```

## Exercise 2

```python
fig = plt.figure()
plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('Y')
axes1 = plt.axes([.25,.6,.25,.22]) 
axes1 = plt.ylabel('Y')
axes1 = plt.xlabel('X')
axes1 = plt.plot(x,y)
```

## Exercise 3



**ใช้  arrays x, y และ z เพื่อทำการ plot บนแกนที่สร้างจากข้อที่แล้ว  (Notice อย่าลืมกำหนด  x - limits และ y - limits )**

```python
fig = plt.figure()
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(x,z)
axes2 = fig.add_axes([.2,.6,.3,.3])
axes2.plot(x,y,y)
axes2.set_ylim(30,55)
axes2.set_xlim(20,22)
```
