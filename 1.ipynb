```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
dic = {"day":['Sun','Mon','Tue','Wed','Thu','Fri','Sat'],"temp":[24,25,29,35,23,20,23]}
df = pd.DataFrame(dic)
x = df['day']
y = df['temp']
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,ls=":",lw ='4')
```

```python
plt.plot(x,y,ls="-",lw ='4')
```

```python
plt.plot(x,y,color ='red')
```

```python
plt.plot(x,y,color =(0,1,0))
```

```python
plt.plot(x,y,color =(0.5,0.5,0))
```

```python
plt.plot(x,y,marker = '<')
```

```python
plt.plot(x,y,marker = '*')#x,p,<,>,*,h,o,s
```

```python
plt.plot(x,y,marker='s',ms=10,mec='k',mfc = 'red')  
                        #ขนาดจุด ขอบสี    สีในจุด
```

```python
plt.plot(x,y,'s:r',lw= 3,mec='k')
```

```python
plt.title("Temperature around this week")
plt.plot(x,y,'s:r',lw= 3,mec='k')
plt.xlabel('Day')
plt.ylabel('Temp')
plt.xticks(rotation = 90)
plt.xticks([15,20,25,30,35,40])
plt.show()
```
