```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('marvel_box_office.csv',encoding = 'iso-8859-1')
df.info()
```

```python
df.head()
```

```python
sns.scatterplot(data=df,x='Opening Weekend',y='Budget', palette='Set1')
```

```python
model = KMeans(n_clusters=3, random_state=0)
df2 = df[['Opening Weekend','Budget']].dropna()
model.fit(df2)
```

```python
plt.figure(figsize=[6,4])
sns.scatterplot(data=df2,x='Opening Weekend',y='Budget', hue=model.labels_,palette='Set1')
plt.scatter(model.cluster_centers_[:,0],model.cluster_centers_[:,1],color='k',marker='*')
```

```python
model.predict([ [ 200,100 ] ,[ 350 , 150 ]])
```
