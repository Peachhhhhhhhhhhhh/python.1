## Data

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
x = np.arange(0,100)
y = x*2
z = x**2
df = pd.read_csv('Superstore.csv',encoding = 'iso-8859-1')
```

## Exercise 1


```python
plt.plot(x,y)
plt.title("Title")
plt.xlabel('x')
plt.ylabel('y')
```

## Exercise 2

```python
fig = plt.figure()
plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('Y')
axes1 = plt.axes([.25,.6,.25,.22]) 
axes1 = plt.ylabel('Y')
axes1 = plt.xlabel('X')
axes1 = plt.plot(x,y)
```

## Exercise 3



**ใช้  arrays x, y และ z เพื่อทำการ plot บนแกนที่สร้างจากข้อที่แล้ว  (Notice อย่าลืมกำหนด  x - limits และ y - limits )**

```python
fig = plt.figure()
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(x,z)
axes1= plt.xlabel('X')
axes1 = plt.ylabel('Z')
axes2 = fig.add_axes([.2,.6,.3,.3])
axes2.plot(x,y,y,color = '#7AAED2' )
axes2.set_ylim(30,55)
axes2.set_xlim(20,22)
axes2= plt.xlabel('X')
axes2 = plt.ylabel('Y')
```

## Exercise 4

**จงใช้คำสั่ง plt.subplots(nrows=1, ncols=2)**

**จากนั้นให้ทำการ plot (x,y) และ plot (x,z) บนแกน axes  และให้ใช้งานคำสั่ง linewidth and style เพื่อตกแต่งเส้นของกราฟ**

```python
fig, axes = plt.subplots(nrows=1 , ncols=2)
axes[0].plot(x,y,'--')
axes[1].plot(x,z,'r')
fig.tight_layout();
fig.set_figheight(2)
fig.set
```

#### โค้ดต่อไปนี้ใช้ในสองข้อสุดท้าย

```python
df['Category'].unique()
```

```python
df[ df['Category'] == 'Furniture' ].groupby('Year').sum()['Profit']
```

##### Dictionary of Category

```python
arr_df = {}
for i in range(0,df['Category'].nunique()) :
    arr_df[df['Category'].unique()[i]] = df[ df['Category'] == df['Category'].unique()[i] ].groupby('Year').sum()['Profit']
```

## จงแสดงกราฟรายได้ของแต่ละ Category ในแต่ละปีในกราฟเดียว 

```python
plt.plot(arr_df["Furniture"].index,arr_df["Furniture"])
```

```python
plt.plot(arr_df["Furniture"].index,arr_df["Furniture"],'^-',label = "Furniture")
plt.plot(arr_df["Office Supplies"].index,arr_df["Office Supplies"],'o-',label = "Office Supplies")
plt.plot(arr_df['Technology'].index,arr_df["Technology"],'s-',label = "Technology")

plt.legend(loc = "best")
```

## จงแสดงกราฟรายได้ของแต่ละ Category ในแต่ละปี แบบแยกกราฟ

```python
fig = plt.figure()
axes1 = fig.add_axes([0.5,1.2,1,1])
axes1.plot(arr_df["Technology"].index,arr_df["Technology"],"s:g",label="Technology")
axes1.set_title("Technology")
axes1.set_xticks(np.arange(2014,2018,1))
axes2 = fig.add_axes([0,0,1,1])
axes2.plot(arr_df["Furniture"].index,arr_df["Furniture"],"^--b",label="Furniture")
axes2.set_title("Furniture")
axes2.set_xticks(np.arange(2014,2018,1))
axes3 = fig.add_axes([1.1,0,1,1])
axes3.plot(arr_df["Office Supplies"].index,arr_df["Office Supplies"],"s-.r",label="Office Supplies")
axes3.set_title("Office Supplies")
axes3.set_xticks(np.arange(2014,2018,1))
```
