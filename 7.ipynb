```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
plt.figure(figsize=[4,2])
sns.histplot(tips["tip"],bins=30,kde = True)
```

```python
plt.figure(figsize=[4,2])
sns.histplot(tips["total_bill"],bins=20,kde = True)
```

```python
plt.figure(figsize=[3,2])
sns.countplot(data=tips,x = "tip")
```

```python
plt.figure(figsize=[3,2])
sns.barplot(data=tips,x="sex",y="tip",)
```

```python
plt.figure(figsize=[3,2])
sns.countplot(data=tips,x = "sex",palette='magma')
```

```python
plt.figure(figsize=[3,2])
sns.boxplot(data=tips,x='day',y='tip')
```

```python
plt.figure(figsize=[4,2])
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='sex')
```

```python
g = sns.JointGrid(data= tips,x='total_bill',y='tip')
g.plot(sns.scatterplot,sns.violinplot)
```

```python
g = sns.PairGrid(data = tips, hue = "sex")
g.map_diag(sns.histplot)
g.map_upper(sns.boxplot)
g.map_lower(sns.violinplot)
```

```python
sns.lmplot(data=tips,x='total_bill',y='tip',col='sex',row = "smoker")
```

```python
g=sns.FacetGrid(data=tips,col='sex',row = "smoker")
g.map(plt.scatter,'total_bill','tip')
```
