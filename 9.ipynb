```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression 
```

```python
df = pd.read_csv("umbrellas_sold_1.csv")
df.head()
```

```python
plt.figure(figsize=(4,3))
plt.scatter(df["rainfall_mm"],df["umbrellas_sold"])
```

```python
att = df[["rainfall_mm"]]
label = df["umbrellas_sold"]
mobel  = LinearRegression()
mobel.fit(att,label)
```

```python
a = mobel.intercept_
b = mobel.coef_
#Y = a + bx
#Y = label,X =att
```

```python
ans = mobel.predict(att)
plt.figure(figsize=(4,3))
plt.scatter(df["rainfall_mm"],df["umbrellas_sold"])
plt.plot(att,ans,color="red")
```

```python
mobel.score(att,label)
```

```python
df2 = pd.read_csv("electricity-consumptions.csv")
att2 = df2[["home_size"]]
label2 = df2["kilowatt_hours_per_month"]
mobel2  = LinearRegression()
mobel2.fit(att2,label2)
ans2 = mobel2.predict(att2)
plt.figure(figsize=(4,3))
plt.scatter(df2["home_size"],df2["kilowatt_hours_per_month"])
plt.plot(att2["home_size"],ans2,'r--')
 
mobel2.score(att2,label2)
```
