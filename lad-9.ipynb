```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
df = sns.load_dataset("mpg")
model = KMeans(n_clusters=4, random_state=0)
df2 = df[['acceleration','weight']].dropna()
model.fit(df2)
plt.figure(figsize=[6,4])
sns.scatterplot(data=df2,x='acceleration',y='weight', hue=model.labels_,palette='Set1')
plt.scatter(model.cluster_centers_[:,0],model.cluster_centers_[:,1],color='k',marker='*')
```

```python
df = sns.load_dataset("mpg")
model = KMeans(n_clusters=3, random_state=0)
df2 = df[['acceleration','weight']].dropna()
scaler = StandardScaler()
scaler.fit(df2)
df2[['acceleration','weight']] = scaler.transform(df2)
model.fit(df2)
plt.figure(figsize=[4,2])
sns.scatterplot(data=df2,x='acceleration',y='weight', hue=model.labels_,palette='Set1')
plt.scatter(model.cluster_centers_[:,0],model.cluster_centers_[:,1],color='k',marker='*')
```
